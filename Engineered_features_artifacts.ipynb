```python

import pandas as pd

# File paths
train_file_path = "C:/Users/joash/Desktop/rr/train_normalized.csv"
test_file_path = "C:/Users/joash/Desktop/rr/test_normalized.csv"

# Function to load, rename columns, and save the dataset
def process_and_save(file_path):
    # Load the dataset
    df = pd.read_csv(file_path)

    # Rename columns: replace & with 'and', remove spaces, and replace commas with underscores
    df.columns = [col.replace('&', 'and').replace(',', '_').replace(' ', '') for col in df.columns]

    # Save the modified dataset back to the same location
    df.to_csv(file_path, index=False)

# Process and save both datasets
process_and_save(train_file_path)
process_and_save(test_file_path)

```
